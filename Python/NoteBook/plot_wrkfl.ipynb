# Using matplot lib

```python
from matplotlib import pyplot as plt

ages_x=list(range(25,36,1))
print(ages_x)

dev_y= [38496,42000,46752,49320,53200,56000,62316,64928,67317,68748,73752]
plt.plot(ages_x,dev_y)

py_dev_y= [45372,48876,54000,58000,61000,65000,70000,75000,77000,80000,83000]
plt.plot(ages_x,py_dev_y)

plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary by Age')

x=plt.show()
```

### Legend the plot using legend with a list of legends for each y axis data set

```python
#First using legent with list of labels for each data set
dev_y= [38496,42000,46752,49320,53200,56000,62316,64928,67317,68748,73752]
plt.plot(ages_x,dev_y)

py_dev_y= [45372,48876,54000,58000,61000,65000,70000,75000,77000,80000,83000]
plt.plot(ages_x,py_dev_y)

plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary by Age')

#Her is rthe legend
plt.legend(['All Dev','Python Dev'])

x=plt.show()
```

### Using labels on each plot and the calling legend

```python
#First using legent with list of labels for each data set
dev_y= [38496,42000,46752,49320,53200,56000,62316,64928,67317,68748,73752]
plt.plot(ages_x,dev_y,label='All Dev')

py_dev_y= [45372,48876,54000,58000,61000,65000,70000,75000,77000,80000,83000]
plt.plot(ages_x,py_dev_y,label='Python Dev')

plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary by Age')

plt.legend()

x=plt.show()
```

### Modifying line style and colors and adding marker to the points

```python
#First using legent with list of labels for each data set
dev_y= [38496,42000,46752,49320,53200,56000,62316,64928,67317,68748,73752]
plt.plot(ages_x,dev_y,label='All Dev',color='k',linestyle='--',marker='o')

py_dev_y= [45372,48876,54000,58000,61000,65000,70000,75000,77000,80000,83000]
plt.plot(ages_x,py_dev_y,label='Python Dev',color='b',marker ='o')

plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary by Age')

plt.legend()

x=plt.show()
```

### Increasing linewithd and adding a new set of data

```python
#First using legent with list of labels for each data set
dev_y= [38496,42000,46752,49320,53200,56000,62316,64928,67317,68748,73752]
plt.plot(ages_x,dev_y,label='All Dev',color='k',linestyle='--',marker='o',linewidth = 1)

py_dev_y= [45372,48876,54000,58000,61000,65000,70000,75000,77000,80000,83000]
plt.plot(ages_x,py_dev_y,label='Python Dev',color='b',marker ='o',linewidth = 2)

java_dev= [38496,42300,46587,50000,53000,56897,62316,56000,67000,68000,74000]
plt.plot(ages_x,java_dev,label='Java Dev',color='r',marker ='.',linewidth = 1)

plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary by Age')

plt.legend()
plt.tight_layout()

#Adding a grid
plt.grid(True)

x=plt.show()
```
